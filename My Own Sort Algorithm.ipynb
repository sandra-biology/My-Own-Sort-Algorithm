```python
import time
import random

def bubblesort(lst):
    """
    Function takes a list and sorts it.
    """
    n = 0
    n_last_to_compare = len(lst) - 1
    # print(f"n_last_to_compare = {n_last_to_compare}")
    
    while n < n_last_to_compare:
        swap = False
        for i in range(n_last_to_compare):
            # print(f"i = {i}, i+1 = {i+1}")
            if lst[i] > lst[i+1]:
                lst[i], lst[i+1] = lst[i+1], lst[i]
                swap = True
        if swap == False:
            break
        n_last_to_compare -= 1

    return lst

N = range(10000)
lst1 = random.choices(N, k=1000)
lst2 = random.choices(N, k=2000)

start1 = time.time()
print(bubblesort(lst1))
end1 = time.time()
time1 = end1 - start1

start2 = time.time()
print(bubblesort(lst2))
end2 = time.time()
time2 = end2 - start2
print(f"time1 = {time1}, time2 = {time2}")
print(f"ratio of times = {time2/time1}")
```
